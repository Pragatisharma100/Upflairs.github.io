```python
city_list=['Ahmedabad', 'Aizawl', 'Amaravati', 'Amritsar', 'Bengaluru', 'Bhopal',
       'Brajrajnagar', 'Chandigarh', 'Chennai', 'Coimbatore', 'Delhi',
       'Ernakulam', 'Gurugram', 'Guwahati', 'Hyderabad', 'Jaipur',
       'Jorapokhar', 'Kochi', 'Kolkata', 'Lucknow', 'Mumbai', 'Patna',
       'Shillong', 'Talcher', 'Thiruvananthapuram', 'Visakhapatnam']
dt={}
for i in range(0,len(city_list)):
    dt[city_list[i]]=i+1
```

```python
df['city']=df['city'].map(dt)
```

```python
ls=list(df['date'])
df_date={}
for i in range(len(ls)):
    if ls[i].endswith('2020'):
         df_date[ls[i]]=2020
    elif ls[i].endswith('2021'):
         df_date[ls[i]]=2021
    elif ls[i].endswith('2022'):
         df_date[ls[i]]=2022
    elif ls[i].endswith('2023'):
         df_date[ls[i]]=2023
```

```python
df['date']=df['date'].map(df_date)
```

```python
aqi_5=df[df['aqi']==5]
aqi_4=df[df['aqi']==4]
aqi_3=df[df['aqi']==3]
aqi_2=df[df['aqi']==2]
aqi_1=df[df['aqi']==1]
```

```python
aqi_5=aqi_5.sample(6000)
```

```python
final_df=pd.concat([aqi_5,aqi_1,aqi_2,aqi_3,aqi_4],axis=0)
```

```python
final_df['aqi'].value_counts()
```

```python
x=final_df.drop('aqi',axis=1)
y=final_df[['aqi']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dfc=DecisionTreeClassifier()
```

```python
dfc.fit(x_train,y_train)
```

```python
print('Training score:',dfc.score(x_train,y_train)*100)
print('Testing score:',dfc.score(x_test,y_test)*100)
```

```python
prediction=dfc.predict(x_test)
```

```python
y_test['prediction']=prediction
```
