IMPORTING THE PANDAS,NUMPY,MATPLOTLIB,SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING THE DATASET OF TIPS

```python
df =sns.load_dataset("tips")  #datframe in seaborn
df.head()
```

CHANGING THE NAME OF DIFFERENT COLUMNS


```python
df.columns=['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'member']
df
```

MAKING A SCATTERPLOT OF TOTAL_BILL AND TIP

```python
# sns.lineplot()
# sns.boxplot()
# sns.scatterplot()
# sns.histplot()
# sns.countplot()
# sns.barplot()
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.show()
```

RELPLOT-> RELATIONAL PLOT (NUMERICAL DATA COLUMNS)      CATPLOT->CATEGORICAL PLOT  (ONE CATEGORIAL AND ONE NUMERICAL DATA COLUMNS)

MAKING A RELPLOT OF TOTAL_BILL AND TIP

```python
sns.relplot(x="total_bill",y="tip",data=df)  
plt.show()
```

RELPLOT THE KIND AS LINE

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="line")    #kind: Literal['scatter', 'line'] = "scatter" 
plt.show()
```

HUE IN RELPLOT

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender")    #hue-> vector or key in data
# Grouping variable that will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.
plt.show()
```

STYLE IN RELPLOT

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker")  #vector or key in data
# Grouping variable that will produce points with different markers. Can have a numeric dtype but will always be treated as categorical.
plt.show()
```

COL IN RELPLOT

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker",col="time")  #DIVIDING IN COLUMNS
plt.show()
```

INSERT DAY DATA IN COL

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",col="day")  
plt.show()
```

FILTER THE DATA OF FEMALE

```python
female=df[df["gender"]=="Female"]
```

FILTER THE DATA OF FEMALE

```python
sns.relplot(x="tip",y="total_bill",data=female)
plt.show()
```

PLOT THE GRAPH AS HUE-SMOKER AND COL AS DAY

```python
sns.relplot(x="tip",y="total_bill",data=female,hue="smoker",col="day")
plt.show()
```

PLOT THE GRAPH AS HUE-GENDER AND ROW-DAY


```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",row="day")  
plt.show()
```
